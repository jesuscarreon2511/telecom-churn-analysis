```
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```
# 1. Cargar dataset
url = "https://raw.githubusercontent.com/ingridcristh/challenge2-data-science-LATAM/main/TelecomX_Data.json"
df = pd.read_json(url)
```

```
# 2. Mostrar primeras filas
print("Primeras filas del dataset:")
print(df.head())
```

```
# 3. Información del dataset
print("\nInformación del dataset:")
print(df.info())
```

```
# 4. Estadísticas descriptivas
print("\nEstadísticas descriptivas:")
print(df.describe())
```

```
# 5. Mostrar nombres de columnas
print("\nColumnas del dataset:")
print(df.columns)
```

```
# 6. Revisar valores nulos
print("\nValores nulos por columna:")
print(df.isnull().sum())
```

```
# 7. Revisar registros duplicados
print("\nNúmero de duplicados:")

print(df.select_dtypes(exclude="object").duplicated().sum())
```

```
# 8. Valores únicos de la variable objetivo
print("\nValores únicos de Churn:")
print(df["Churn"].unique())
```

```
# 9. Convertir nombres de columnas a minúsculas
df.columns = df.columns.str.lower()
print("\nColumnas en minúsculas:")
print(df.columns)
```

```
# 10. Transformar variable objetivo a numérica
df["churn"] = df["churn"].map({
    "Yes": 1,
    "No": 0
})
```

```
print("\nDistribución de churn (0 = No, 1 = Yes):")
print(df["churn"].value_counts())
```
